```python
# Building dataframe for sectors
df_sector = df_users_with_duplicates[['Nombre y Apellido', 'Estados_de_USA']]
df_sector['sector'] = df[['En que rea principalmente desarrollas tu profesin? (Si quieres ingresar ms de una opcin por favor ingresa las opciones en el campo "Otros")']]

# df_sector = df[['Nombre y Apellido', 'Estados_de_USA', 'En que rea principalmente desarrollas tu profesin? (Si quieres ingresar ms de una opcin por favor ingresa las opciones en el campo "Otros")']]
# df_sector['Estados_de_USA'] = df_users_with_duplicates['Estados_de_USA']
# df_sector = df_sector.rename({'En que rea principalmente desarrollas tu profesin? (Si quieres ingresar ms de una opcin por favor ingresa las opciones en el campo "Otros")':'sector'},axis=1)
```

```python
#@title
print('Cantidad de celdas con valores "null":\n ')
df_sector.isna().sum()
```

```python
#@title
df_sector = df_sector.drop_duplicates()
len(df)-len(df_sector)
print('Cantidad de celdas duplicadas eliminadas: ' + str(len(df)-len(df_sector)))
```

```python
# Original options in the survey.
list_sector= ['Medicina Laboral',
              'Tratamiento del lenguaje en nios',
              'Tratamiento del lenguaje en adultos',
              'Audiologa en adultos',
              'Audiologa en nios',
              'Adaptacin de prtesis auditivas']
# 'Docencia' is added after analyzing the data
list_sector.append('Docencia')
df_sector[list_sector] = 0

for i,item in enumerate(df_sector['sector']):
    string = item.lower()
    string = string.replace(' en',' ').replace(' del',' ').replace(' de',' ').replace(' y',',').replace('  ',' ')
    
    # When 'Laboral' is mentioned (i.e. Laboral , rehabilitacin nios y adulto , audfonos , Audiologa	)
    if 'medicina laboral' in string or 'laboral' in string:
        df_sector['Medicina Laboral'].iloc[i]=1
    # When 'Prtesis' is mentioned (i.e. Audiologa y adaptacin de prtesis auditivas	)
    if 'prt' in string or 'prot' in string or 'audfono' in string or 'audifono' in string:
        df_sector['Adaptacin de prtesis auditivas'].iloc[i]=1
    # When 'Docencia' is in mentioned(i.e. Audiologia clinica en adultos, nios y protesis auditivas y docencia en la universodad catolica de salta en la materia terapeutica audiologica	)
    if 'docencia' in string:
        df_sector['Docencia'].iloc[i]=1
    # When 'nio' is mentioned, it is parsed to understand if it is connected to 'Audiologa' or 'Tratamiento del lenguaje' (i.e.'Audiologa en nios'; 'Audiologia adultos, nios, laboral y tratamiento delLenguaje adultos y nios')
    if 'os' in string:
        if 'trat' in string.split('os')[0] or 'rehab' in string.split('os')[0] or 'leng' in string.split('os')[0]:
            df_sector['Tratamiento del lenguaje en nios'].iloc[i]=1
        elif len(string.split('os'))>2:
            if 'trat' in string.split('os')[1] or 'rehab' in string.split('os')[1] or 'leng' in string.split('os')[1]:
                df_sector['Tratamiento del lenguaje en nios'].iloc[i]=1
        if 'audiologa' in string.split('os')[0] or 'audiologia' in string.split('os')[0] or 'audio ' in string.split('os')[0]:
            df_sector['Audiologa en nios'].iloc[i]=1
        elif len(string.split('os'))>2:
            if 'audiologa' in string.split('os')[1] or 'audiologia' in string.split('os')[1] or 'audio ' in string.split('os')[1] or 'audiometria' in string.split('os')[1] or 'audiometra' in string.split('os')[1]:
                df_sector['Audiologa en nios'].iloc[i]=1
        # When it is called 'nios' as area of expertise, but it is not specified whether it corresponds to 'Audiologa' or 'Tratamiento del lenguaje'
        if string=='nios':
            df_sector['Tratamiento del lenguaje en nios'].iloc[i]=1
            df_sector['Audiologa en nios'].iloc[i]=1
    # When 'adulto' is mentioned, it is parsed to understand if it is connected to 'Audiologa' or 'Tratamiento del lenguaje' (i.e.'Audiologa en adultos'; 'Audiologia adultos, nios, laboral y tratamiento delLenguaje adultos y nios')
    if 'adult' in string:
        if 'trat' in string.split('adult')[0] or 'rehab' in string.split('adult')[0] or 'leng' in string.split('adult')[0]:
            df_sector['Tratamiento del lenguaje en adultos'].iloc[i]=1
        elif len(string.split('adult'))>2:
            if 'trat' in string.split('adult')[1] or 'rehab' in string.split('adult')[1] or 'leng' in string.split('adult')[1]:
                df_sector['Tratamiento del lenguaje en adultos'].iloc[i]=1
        if 'audiologa' in string.split('adult')[0] or 'audiologia' in string.split('adult')[0] or 'audio ' in string.split('adult')[0]:
            df_sector['Audiologa en adultos'].iloc[i]=1
        elif len(string.split('adult'))>2:
            if 'audiologa' in string.split('adult')[1] or 'audiologia' in string.split('adult')[1] or 'audio ' in string.split('adult')[1]:
                df_sector['Audiologa en adultos'].iloc[i]=1
    # When there are 'adulto' or 'nio' mentioned but no for 'AUDIOLOGA' that stay not specified (i.e. 'Lenguaje de nios y audiologa en todas las edades').
    if 'adult' in string or 'os' in string: 
        if len(string.split('os')[len(string.split('os'))-1]) > len(string.split('adult')[len(string.split('adult'))-1]):
            if 'audiologa' in string.split('adult')[len(string.split('adult'))-1] or 'audiologia' in string.split('adult')[len(string.split('adult'))-1] or 'audio ' in string.split('adult')[len(string.split('adult'))-1]:
                df_sector['Audiologa en adultos'].iloc[i]=1
                df_sector['Audiologa en nios'].iloc[i]=1
        else:
            if 'audiologa' in string.split('os')[len(string.split('os'))-1] or 'audiologia' in string.split('os')[len(string.split('os'))-1] or 'audio ' in string.split('os')[len(string.split('os'))-1] or 'audiometria' in string.split('os')[len(string.split('os'))-1] or 'audiometra' in string.split('os')[len(string.split('os'))-1]:
                df_sector['Audiologa en nios'].iloc[i]=1
                df_sector['Audiologa en adultos'].iloc[i]=1
    # When there are no 'adulto' or 'nio' mentioned, so 'AUDIOLIGA' is not specified(i.e. 'Audiologa y adaptacin de prtesis auditivas').
    if not('adult' in string) and not('os' in string): 
        if 'audiologa' in string or 'audiologia' in string:
            df_sector['Audiologa en adultos'].iloc[i]=1
            df_sector['Audiologa en nios'].iloc[i]=1
```

Las dos!
Se puede decir que las frecuencias relativas son ms tiles a la hora de tomar decisiones comparando grupos de datos.
Ya que considerar que una situacin "A" se da ms frecuentemente que otra "B" (en funcin del total de los datos y dependiendo de otra variable) indicara que la correlacin tiene tendencia por la situacin A.

Sin embargo, la cantidad de datos es sumamente importante. Si el 75% de los datos indican una tendencia, pero slo tengo 4 datos en total esta tendencia tiene algn sentido? En la vida real: NO.

> Por esta razn, nunca hay que perder de vista las frecuencias absolutas cuando se comparan las frecuencias relativas.
